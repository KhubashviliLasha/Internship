```python
import pandas as pd

# Read the Excel file
sheets = pd.read_excel(r'D:\Grenoble_university\6_month_internship\working space\data\combined_1_year_data_with_flexibility.xlsx', sheet_name=None)

# Initialize an empty dictionary to store results from each sheet
hourly_sums_all_sheets = {}

# Iterate through each sheet
for sheet_name, df in sheets.items():
    # Find the column with "Washing machine" in its name
    washing_machine_col = None
    for col in df.columns:
        if "Washing machine" in col:
            washing_machine_col = col
            break
    
    # Check if the column was found
    if washing_machine_col is None:
        print(f"No 'Washing machine' column found in sheet {sheet_name}")
        continue
    
    # Filter rows to include only the washing machine data
    washing_machine_df = df[[washing_machine_col, 'date']].copy()

    # Convert 'date' column to datetime format
    washing_machine_df.loc[:, 'date'] = pd.to_datetime(washing_machine_df['date'])

    # Define the time ranges for calculation (excluding 8 am - 10 am and 6 pm - 10 pm)
    excluded_ranges = [
        (pd.Timestamp('08:00:00').time(), pd.Timestamp('10:00:00').time()),
        (pd.Timestamp('18:00:00').time(), pd.Timestamp('22:00:00').time())
    ]

    # Calculate sum for each hour outside the excluded ranges
    hourly_sums = {}
    for hour in range(24):
        if any(start <= pd.Timestamp(f'{hour}:00:00').time() < end for start, end in excluded_ranges):
            continue  # Skip the excluded time ranges
        start_time = pd.Timestamp(f'{hour}:00:00').time()
        end_time = pd.Timestamp(f'{hour}:59:59').time()
        hourly_sum = washing_machine_df[
            (washing_machine_df['date'].dt.time >= start_time) & 
            (washing_machine_df['date'].dt.time <= end_time)
        ][washing_machine_col].sum()
        hourly_sums[f'{hour}:00 - {hour}:59'] = hourly_sum

    # Store the results for this sheet
    hourly_sums_all_sheets[sheet_name] = hourly_sums

# Print the sum for the washing machine for each hour and each sheet
for sheet_name, hourly_sums in hourly_sums_all_sheets.items():
    print(f"Sheet: {sheet_name}")
    for hour, sum_ in hourly_sums.items():
        print(f"Washing machine for {hour}: {sum_}")
    print("\n")
```

```python
import pandas as pd

# Read the Excel file and load all sheets
dfs = pd.read_excel(r'D:\Grenoble_university\6_month_internship\working space\data\combined_1_year_data_with_flexibility.xlsx', sheet_name=None)

# Specify the sheets you want to process
sheet_names = ['2000993', '2000968', '2000983', '2000915']

# Initialize a dictionary to store hourly sums for each sheet and device
hourly_sums_by_sheet = {}

# Loop through each specified sheet
for sheet_name in sheet_names:
    df = dfs[sheet_name]

    # Convert 'date' column to datetime format
    df['date'] = pd.to_datetime(df['date'])

    # Drop the first two columns ("Weekday" and "date") to get the device columns
    device_columns = df.columns[2:]

    # Initialize a dictionary to store hourly sums for each device in the current sheet
    hourly_sums_by_device = {}

    # Loop through each device column
    for device_col in device_columns:
        # Filter rows to include only the device data
        device_df = df[device_col]

        # Calculate the sum for the device for each hour of the day
        hourly_sums = {}
        for hour in range(24):
            start_time = pd.Timestamp(f'{hour}:00:00').time()
            end_time = pd.Timestamp(f'{hour}:59:59').time()
            hourly_sum = device_df[
                (df['date'].dt.time >= start_time) & 
                (df['date'].dt.time <= end_time)
            ].sum()
            hourly_sums[f'{hour}:00 - {hour}:59'] = hourly_sum
        
        # Store the hourly sums for the device in the dictionary
        hourly_sums_by_device[device_col] = hourly_sums

    # Store the hourly sums for all devices in the current sheet
    hourly_sums_by_sheet[sheet_name] = hourly_sums_by_device

# Print the sum for each device for each hour for each sheet
for sheet_name, device_sums in hourly_sums_by_sheet.items():
    print(f"Hourly sums for sheet {sheet_name}:")
    for device, hourly_sums in device_sums.items():
        print(f"Device: {device}")
        for hour, sum_ in hourly_sums.items():
            print(f"  {hour}: {sum_}")
    print("\n")  # Add a newline for better readability
```

```python
import pandas as pd

# Read the Excel file
df = pd.read_excel('D:\Grenoble_university\6_month_internship\working space\data\combined_1_year_data_with_flexibility.xlsx', sheet_name=None)

# Filter rows to include only the washing machine (assuming it's in a specific column, adjust as needed)
washing_machine_df = df['Washing machine ()']

# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Initialize total consumption
total_consumption = 0

# Calculate the total consumption for the washing machine for each hour of the day
for hour in range(24):
    start_time = pd.Timestamp(f'{hour}:00:00').time()
    end_time = pd.Timestamp(f'{hour}:59:59').time()
    hourly_sum = washing_machine_df[
        (df['date'].dt.time >= start_time) & 
        (df['date'].dt.time <= end_time)
    ].sum()
    total_consumption += hourly_sum

# Print the total consumption for the washing machine
print(f"Total consumption for the washing machine: {total_consumption}")
```
