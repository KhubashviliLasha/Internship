```python
import pandas as pd

# Read the column titles dataframe
file_path = r'D:\Grenoble_university\6_month_internship\working space\flexibility indicators\column_titles_all.xlsx'
column_titles = pd.read_excel(file_path)

# Filter out rows corresponding to non-appliance entries
column_titles = column_titles[~column_titles['Names of household appliances'].str.contains('Site consumption|Total site light consumption')]

# Define a dictionary to store the flexibility values for each appliance
flexibility_mapping = {}

# Iterate over appliance names
for appliance in column_titles['Names of household appliances']:
    # Split the appliance name into words
    words = appliance.split(' ')
    
    # Check if the appliance name has more than one word
    if len(words) > 1:
        # Join the first two words as the prefix
        prefix = ' '.join(words[:5])
    else:
        # Consider the first word as the prefix
        prefix = words[0]
        
    # If the last character of the prefix is '(', remove it
    if '(' in prefix:
        # Remove anything after the last occurrence of parentheses
        prefix = prefix[:prefix.rfind('(')]
    if '1' in prefix:
        # Remove anything after the last occurrence of 1 
        prefix = prefix[:prefix.rfind('1')]
    if '2' in prefix:
        # Remove anything after the last occurrence of 2
        prefix = prefix[:prefix.rfind('2')]
    if '3' in prefix:
        # Remove anything after the last occurrence of 3
        prefix = prefix[:prefix.rfind('3')]
    if '4' in prefix:
        # Remove anything after the last occurrence of 4
        prefix = prefix[:prefix.rfind('4')]
    if '5' in prefix:
        # Remove anything after the last occurrence of 5
        prefix = prefix[:prefix.rfind('5')]
    # Check if there are parentheses in the prefix
    if '(' in prefix:
        # Remove anything after the last occurrence of parentheses
        prefix = prefix[:prefix.rfind('(')]
    
    # Check if there are any numbers in the prefix
    if any(char.isdigit() for char in prefix):
        prefix = prefix[:prefix.rfind(' ')]
    
    # If the prefix is empty, use the entire appliance name as the prefix
    if not prefix.strip():
        prefix = appliance
    
    # Check if the prefix is already in the flexibility_mapping dictionary
    if prefix not in flexibility_mapping:
        # If not, add it with the flexibility value
        flexibility_mapping[prefix] = column_titles.loc[column_titles['Names of household appliances'] == appliance, 'Flexibility'].iloc[0]

# Convert flexibility_mapping dictionary to DataFrame
flexibility_df = pd.DataFrame.from_dict(flexibility_mapping, orient='index', columns=['Flexibility'])

# Reset index to make the prefixes a regular column
flexibility_df.reset_index(inplace=True)
flexibility_df.rename(columns={'index': 'Appliance'}, inplace=True)
```

```python
import pandas as pd

# Read the dataframe for the specific sheet
sheet_name = '2000997'  # Change this to the desired sheet name e.i house ID 
df = pd.read_excel(r'D:\Grenoble_university\6_month_internship\working space\data\filtered_data_for_1_year.xlsx', sheet_name=sheet_name)

# Create a dictionary to store the flexibility values for each column
flexibility_dict = {}

for column_name in df.columns:
    # Extract prefix from the column name
    words = column_name.split(' ')
    prefix_matched = False
    for appliance in column_titles['Appliance']:
        # Check if at least one word from the column name matches a word in the appliance name
        if any(word in appliance for word in words):
            prefix_matched = True
            flexibility_dict[column_name] = column_titles.loc[column_titles['Appliance'] == appliance, 'Flexibility'].iloc[0]
            break
    
    # If no matching appliance found, mark flexibility as 'Uncategorized'
    if not prefix_matched:
        flexibility_dict[column_name] = 'Uncategorized'
    
    # Check if the column name is 'Total site light consumption ()'
    if 'Total site light consumption ()' in column_name:
        flexibility_dict[column_name] = 'Uncategorized'

# Add flexibility values as the first row of the dataframe
flexibility_row = pd.Series(flexibility_dict, name='Flexibility')
df_with_flexibility = pd.concat([flexibility_row.to_frame().T, df], ignore_index=False)

# Set 'Weekday' as the index
df_with_flexibility.set_index('Weekday', inplace=True)

# Reset the index to move 'Weekday' back as a regular column
df_with_flexibility.reset_index(inplace=True)

# Replace "Uncategorized" with an empty string ('') for the "Weekday" column
df_with_flexibility['Weekday'] = df_with_flexibility['Weekday'].replace('Uncategorized', '')
df_with_flexibility['date'] = df_with_flexibility['date'].replace('Uncategorized', '')


# Save the modified DataFrame to an Excel file
# df_with_flexibility.to_excel(f'modified_combined_data_{sheet_name}.xlsx', index=False)
```
