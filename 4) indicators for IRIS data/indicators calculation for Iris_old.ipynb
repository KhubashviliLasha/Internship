## calculate flexible devices total energy

```python
# Extract flexibility information from the second row
flexibility_row = df.iloc[0, 3:]

# Identify flexible devices based on the flexibility information
flexible_devices = [col for col, flexibility in zip(df.columns[3:], flexibility_row) if flexibility.lower() == 'flexible']

# Filter out the first two rows
df_numeric = df.iloc[1:]

# Calculate the sum of each flexible power
flexible_sum_2000900 = df_numeric[flexible_devices].sum()

# Print the sum of each flexible power
print("Sum of each flexible power:")
print(flexible_sum_2000900)
```

```python
# Calculate the total sum of power consumption for all flexible devices
total_flexible_power_sum = df_numeric[flexible_devices].sum().sum()

# Print the total sum of power consumption for all flexible devices
print("Total sum of power consumption for all flexible devices:", total_flexible_power_sum)
```

```python
# Extract flexibility information from the second row
Not_flexibility_row = df.iloc[0, 3:]

# Identify flexible devices based on the flexibility information
Not_flexible_devices = [col for col, Not_flexibility in zip(df.columns[3:], Not_flexibility_row) if Not_flexibility.lower() == 'not_flexible']

# Filter out the first two rows
df_numeric = df.iloc[1:]

# Calculate the sum of each flexible power
Not_flexible_sum_2000900 = df_numeric[Not_flexible_devices].sum()

# Print the sum of each flexible power
print("Sum of each Not_flexible power:")
print(Not_flexible_sum_2000900)
```

```python
# Calculate the total sum of power consumption for all flexible devices
total_Not_flexible_power_sum = df_numeric[Not_flexible_devices].sum().sum()

# Print the total sum of power consumption for all flexible devices
print("Total sum of power consumption for all Not_flexible devices:", total_flexible_power_sum)
```

# all sheet calculation 

```python
import pandas as pd

# Define the range of sheet names
start_sheet = 2000900
end_sheet = 2000997

# Initialize a dictionary to store the sum of each flexible power for each sheet
flexible_sum_per_sheet = {}

# Iterate through each sheet in the specified range
for sheet_num in range(start_sheet, end_sheet + 1):
    sheet_name = str(sheet_num)
    try:
        # Read the sheet
        df = pd.read_excel("modified_combined_data_all_sheets_final.xlsx", sheet_name=sheet_name)

        # Extract flexibility information from the second row
        flexibility_row = df.iloc[0, 3:]

        # Identify flexible devices based on the flexibility information
        flexible_devices = [col for col, flexibility in zip(df.columns[3:], flexibility_row) if flexibility.lower() == 'flexible']

        # Filter out the first two rows
        df_numeric = df.iloc[1:]

        # Calculate the sum of each flexible power
        flexible_sum = df_numeric[flexible_devices].sum()

        # Store the sum of each flexible power for the current sheet
        flexible_sum_per_sheet[sheet_name] = flexible_sum
    except Exception as e:
        print(f"Error processing sheet {sheet_name}: {e}")

# Print the sum of each flexible power for each sheet
for sheet_name, flexible_sum in flexible_sum_per_sheet.items():
    print(f"Sheet {sheet_name}:")
    print(flexible_sum)


```

# to save as excel file

```python

# Convert the dictionary into a DataFrame
flexible_sum_df = pd.DataFrame(flexible_sum_per_sheet)

# Save the DataFrame to an Excel file
#flexible_sum_df.to_excel("flexible_sum_per_sheet.xlsx")
```

```python
flexible_sum_df
```

```python
# Initialize a dictionary to store the sum of each device across all sheets
total_flexibility_per_device = {}

# Iterate through each sheet
for flexible_sum in flexible_sum_per_sheet.values():
    # Iterate through each device in the current sheet
    for device, value in flexible_sum.items():
        # Add the value to the total for the current device
        total_flexibility_per_device[device] = total_flexibility_per_device.get(device, 0) + value

# Print the sum of flexibility for each device across all sheets
print("Total flexibility sum for each device:")
for device, total_sum in total_flexibility_per_device.items():
    print(f"{device}: {total_sum}")
```

```python
# Initialize the total flexibility sum
global_flexibility_sum = 0

# Sum up all the flexibility values across all devices
for total_sum in total_flexibility_per_device.values():
    global_flexibility_sum += total_sum

# Print the total flexibility sum
print("Total flexibility sum across all devices:", global_flexibility_sum)
```

## not flexible sum 

```python
import pandas as pd

# Define the range of sheet names
start_sheet = 2000900
end_sheet = 2000997

# Initialize a dictionary to store the sum of each not flexible power for each sheet
not_flexible_sum_per_sheet = {}

# Iterate through each sheet in the specified range
for sheet_num in range(start_sheet, end_sheet + 1):
    sheet_name = str(sheet_num)
    try:
        # Read the sheet
        df = pd.read_excel("modified_combined_data_all_sheets_final.xlsx", sheet_name=sheet_name)

        # Extract flexibility information from the second row
        flexibility_row = df.iloc[0, 3:]

        # Identify not flexible devices based on the flexibility information
        not_flexible_devices = [col for col, flexibility in zip(df.columns[3:], flexibility_row) if flexibility.lower() != 'flexible']

        # Filter out the first two rows
        df_numeric = df.iloc[1:]

        # Calculate the sum of each not flexible power
        not_flexible_sum = df_numeric[not_flexible_devices].sum()

        # Store the sum of each not flexible power for the current sheet
        not_flexible_sum_per_sheet[sheet_name] = not_flexible_sum
    except Exception as e:
        print(f"Error processing sheet {sheet_name}: {e}")

# Print the sum of each not flexible power for each sheet
for sheet_name, not_flexible_sum in not_flexible_sum_per_sheet.items():
    print(f"Sheet {sheet_name}:")
    print(not_flexible_sum)
```

```python
# Convert the dictionary into a DataFrame
not_flexible_sum_df = pd.DataFrame(not_flexible_sum_per_sheet)

# Save the DataFrame to an Excel file
#not_flexible_sum_df.to_excel("not_flexible_sum_per_sheet.xlsx")
```

```python
# Initialize the total flexibility sum
global_not_flexibility_sum = 0

# Sum up all the flexibility values across all devices
for total_sum in total_not_flexibility_per_device.values():
    global_not_flexibility_sum += total_sum

# Print the total flexibility sum
print("Total flexibility sum across all devices:", global_not_flexibility_sum)
```
